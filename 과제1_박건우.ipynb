```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
```

```python
html = urlopen('http://dh.aks.ac.kr/Encyves/wiki/index.php/Contents_B')
```

```python
soup = bs(html, 'html.parser')
```

```python
soup #html 파일 참고용
```

```python
table = soup.find_all('table')[0]
trs = table.find_all('tr') # 테이블 일단 다 가져오기
trs # 참고용 데이터 확인하려고
```

```python
names = []
types = []
for tr in trs:
    anchors = tr.select('a')
    if len(anchors) > 0:
        names.append(anchors[1].text)
        types.append(tr.select('td')[2].text)

data = {
    '의궤 종류': types,
    '의궤 이름': names
}

df = pd.DataFrame.from_dict(data)
df[0:10]
```
